# **Importing Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings('ignore')
```

```python
# 1] Loading the Dataset into Pandas DataFrame

credit_card_data = pd.read_csv("/content/drive/MyDrive/Colab datasets/creditcard.csv")
```

```python
# Reading the first 5 rows

credit_card_data.head()
```

```python
# Reading the last 5 rows

credit_card_data.tail()
```

```python
# Dataset Information

credit_card_data.info()
```

```python
# Checking Null values in each column

credit_card_data.isnull().sum()
```

```python
# Distributions of Legit & Fraudulent transaction counts

credit_card_data['Class'].value_counts()
```

```python
# Seperating the data

legit_transactions = credit_card_data[credit_card_data.Class == 0]
fraud_transactions = credit_card_data[credit_card_data.Class == 1]
```

```python
# Statistical values of the data

legit_transactions.Amount.describe()
```

```python
# Statistical values of the data

fraud_transactions.Amount.describe()
```

```python
# Comparing the mean values for both transactions

credit_card_data.groupby('Class').mean()
```

## **Building a sample dataset from the original dataset to make it more balanced.**

```python
# Using random sampling

legit_transactions_sample = legit_transactions.sample(n=492)
```

```python
new_dataset = pd.concat([legit_transactions_sample,fraud_transactions], axis = 0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

# **Splitting the data into features & targets**

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

# **Split the data into Training and Testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(Y.shape, Y_train.shape, Y_test.shape)
```

# **Training the model**

# **LOGISTIC REGRESSION**

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression model using Training data

model.fit(X_train, Y_train)
```

# **Model Evaluation**

```python
# Accuracy score using training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy_percent = round(training_data_accuracy*100, 2)
```

```python
print('Accuracy Score of Training data:', training_data_accuracy_percent)
```

```python
# Accuracy score using testing data

X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
testing_data_accuracy_percent = round(testing_data_accuracy*100, 2)
```

```python
print('Accuracy Score of Training data:', testing_data_accuracy_percent)
```
